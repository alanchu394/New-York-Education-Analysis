```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("middleSchoolData.csv")  
df= df.dropna(subset = ["rigorous_instruction","collaborative_teachers","supportive_environment","effective_school_leadership","strong_family_community_ties","trust","student_achievement","reading_scores_exceed","math_scores_exceed"])
df=(df-df.mean())/df.std()

#calculate overall rating by adding all the rating from columns L-Q and averaging the values
x = df["rigorous_instruction"]+df["collaborative_teachers"]+df["supportive_environment"]+df["effective_school_leadership"]+df["strong_family_community_ties"]+df["trust"]
x = x/6
#calculate overall academic achievement by adding student achievement to reading 
#df["student_achievement"] = df["student_achievement"]/df["student_achievement"].max()

y = df["student_achievement"] + df["reading_scores_exceed"] + df["math_scores_exceed"]
y=y/3
correlation = x.corr(y)
correlation2 = np.corrcoef(x,y)[0][1]
print(correlation)
print(correlation2)
plt.scatter(x,y)

```
