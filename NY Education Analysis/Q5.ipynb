```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import mannwhitneyu


df = pd.read_csv("middleSchoolData.csv") 
df = df.dropna(subset=["rigorous_instruction","student_achievement"])
ave = np.median(df["rigorous_instruction"])


RI = np.array(df["rigorous_instruction"])
SA = np.array(df["student_achievement"])
lowRI= []
highRI= []

for x in range(len(RI)):
    if(RI[x] <= ave):lowRI.append(SA[x])
    else: highRI.append(SA[x])

print(df["rigorous_instruction"].corr(df["student_achievement"]))
t = stats.ttest_ind(lowRI,highRI)

#df.loc[df['rigorous_instruction'] <= ave] = 0
#df.loc[df['rigorous_instruction'] > ave] = 1
#print(np.mean(lowRI))
##print(np.mean(highRI))
#print(np.std(lowRI))
#print(np.std(highRI))
#print(len(lowRI))
#print(len(highRI))

print(t)

#print(df["rigorous_instruction"])
plt.scatter(RI,SA)
```
