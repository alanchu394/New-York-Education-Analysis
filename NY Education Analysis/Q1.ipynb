```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("middleSchoolData.csv")  

np.where(pd.isnull(df["applications"]))
np.where(pd.isnull(df["acceptances"]))

df.plot.scatter("applications","acceptances")


x = df["applications"]
y = df["acceptances"]
correlation = x.corr(y)
print(correlation)
import statsmodels.api as sm
X =x
X = sm.add_constant(X.ravel())
results = sm.OLS(y,X).fit()
results.summary()  
```
